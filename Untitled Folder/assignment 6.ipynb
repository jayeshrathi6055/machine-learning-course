```python
# Step - 3 - Preprocess the data
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
clean_sms_lst = []

sms_len_lst = []
```

```python
def preprocess(raw_sms, flag):
    # Removing special characters and digits
    letters_only = re.sub("[^a-zA-Z]", " ",raw_sms)
    
    # change sentence to lower case
    letters_only = letters_only.lower()

    # tokenize into words
    words = letters_only.split()
    
    # remove stop words                
    words = [w for w in words if not w in stopwords.words("english")]
    
    #Stemming/Lemmatization
    if(flag == 'stem'):
        words = [stemmer.stem(word) for word in words]
    else:
        words = [lemmatizer.lemmatize(word) for word in words]
    
    clean_sms_lst.append(" ".join(words))
    
    sms_len_lst.append(len(words))
```

```python
from tqdm import tqdm, tqdm_notebook
tqdm.pandas()
```

```python
import nltk
nltk.download('wordnet')
```

```python

clean_sms_lst = []

sms_len_lst = []

df['message'].progress_apply(lambda x: preprocess(x, 'stem'))

df['clean_sms_stem'] = clean_sms_lst

df['sms_length_stem'] = sms_len_lst
```

```python
clean_sms_lst = []

sms_len_lst = []

df['message'].progress_apply(lambda x: preprocess(x, 'lema'))

df['clean_sms_lema'] = clean_sms_lst

df['sms_length_lema'] = sms_len_lst
```

```python
df.head()
```

```python
#Step - 4 - Data Preparation
df['target_1']=df['target'].apply(lambda x: 0 if x=='spam' else 1)
df.head()
```

```python
# splitting into test and train

from sklearn.model_selection  import train_test_split

train, test = train_test_split(df,test_size=0.2,random_state=42)
```

```python
train_clean_sms=[]
for sms in train['clean_sms_stem']:
    train_clean_sms.append(sms)

test_clean_sms=[]
for sms in test['clean_sms_stem']:
    test_clean_sms.append(sms)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = "word")

train_features = vectorizer.fit_transform(train_clean_sms)

test_features = vectorizer.transform(test_clean_sms)
```

```python
vectorizer.vocabulary_
```

```python
print("Total unique words:", len(vectorizer.vocabulary_))

print("Type of train_features:", type(train_features))

print("Shape of input data", train_features.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn import metrics
```

```python
Classifiers = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    SVC()]
```

```python
dense_features = train_features.toarray()

dense_test = test_features.toarray()

for classifier in tqdm(Classifiers):
    fit = classifier.fit(dense_features,train['target_1'])
    pred = fit.predict(dense_test)
# Step - 6 - Evaluation. For each Model, evaluate using following metrics:
    accuracy = metrics.accuracy_score(pred,test['target_1'])
    print('Accuracy of '+classifier.__class__.__name__+' is '+str(accuracy))
    print(metrics.confusion_matrix(pred,test['target_1']))
```

```python
print(metrics.classification_report(pred,test['target_1']))
```
