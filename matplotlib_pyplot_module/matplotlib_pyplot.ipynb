# Matplotlib.pyplot Tutorial

```python
# Import Module
import matplotlib.pyplot as plt
```

## Matplotlib Plotting

```python
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

# Plot with line
plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

# Plot without line
plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

## Matplotlib Markers

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*')
plt.show()
```

```python
# Marker	Description

# 'o'	Circle	
# '*'	Star	
# '.'	Point	
# ','	Pixel	
# 'x'	X	
# 'X'	X (filled)	
# '+'	Plus	
# 'P'	Plus (filled)	
# 's'	Square	
# 'D'	Diamond	
# 'd'	Diamond (thin)	
# 'p'	Pentagon	
# 'H'	Hexagon	
# 'h'	Hexagon	
# 'v'	Triangle Down	
# '^'	Triangle Up	
# '<'	Triangle Left	
# '>'	Triangle Right	
# '1'	Tri Down	
# '2'	Tri Up	
# '3'	Tri Left	
# '4'	Tri Right	
# '|'	Vline	
# '_'	Hline
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

```python
# Line Syntax	Description
# '-'	Solid line	
# ':'	Dotted line	
# '--'	Dashed line	
# '-.'	Dashed/dotted line
```

```python
# Color Syntax	Description
# 'r'	Red	
# 'g'	Green	
# 'b'	Blue	
# 'c'	Cyan	
# 'm'	Magenta	
# 'y'	Yellow	
# 'k'	Black	
# 'w'	White
```

```python
ypoints = np.array([3, 8, 1, 10])

# Set the marker size
plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

# Set the Edge colour
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

# Set the face colour
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r', mec = 'r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

# Set the colour by hexadecimal code
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'hotpink', mfc = 'hotpink')
plt.show()
```

## Matplotlib Line

```python
ypoints = np.array([3, 8, 1, 10])

# Dotted Line
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
# Dashed Line
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

```python
# Shorter Syntax
plt.plot(ypoints, ls = ':')
plt.show()
```

```python
# Style	Or
# 'solid' (default)	'-'	
# 'dotted'	':'	
# 'dashed'	'--'	
# 'dashdot'	'-.'	
# 'None'	'' or ' '
```

```python
ypoints = np.array([3, 8, 1, 10])

# Set Line width
plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

# Multiple Lines
plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

# Make 2 lines on same graph
plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')   # marker(o)line(:)color(r)
plt.show()
```

## Matplotlib Labels

```python
# Use of xlabel() and ylabel()
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
# Use of title()
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
# Want to change font
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
# Set the Position of the title
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

## Matplotlib Grid

```python
# Use of grid()
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
# Want lines only at x-position
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

```python
# Want lines only at y-position
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(axis = 'y')

plt.show()
```

```python
# Set linestyle and linewidth of grid lines
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 1.5)

plt.show()
```

## Matplotlib Subplot

```python
# Use of subplot()

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)  # (nrows, ncols, plot_number)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)  # (nrows, ncols, plot_number)
plt.plot(x,y)

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2,1,1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2,1,2)
plt.plot(x,y)

plt.show()
```

```python
# Use of suptitle()

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```
